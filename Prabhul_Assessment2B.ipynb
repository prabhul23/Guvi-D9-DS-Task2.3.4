###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df1 = pd.read_csv("college_1.csv")
df2 = pd.read_csv("college_2.csv")
df1.head()
```

```python
df2.head()
```

```python
df3 = pd.concat([df1,df2],axis =0, ignore_index= True)
df3.head(120)
```

```python
df3 = df3.replace([-1],0)
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
ex_df = df3[df3['CodeKata Score'] > 15000]
re_df = df3[(df3['CodeKata Score'] < 15000) & (df1['CodeKata Score'] > 10000)]
ni_df = df3[(df3['CodeKata Score'] < 10000) & (df1['CodeKata Score'] > 7000)]
un_df = df3[(df3['CodeKata Score'] < 7000)]
ex_df.to_csv("Exceeded_Expectation.csv")
re_df.to_csv("Reached_expectation.csv")
ni_df.to_csv("Needs_improvement.csv")
un_df.to_csv("Unsatisfactory.csv")
```

###### Department wise codekata performence (pie chart)

```python
departments = [x for x in df3['Department'].unique()]
ee_df= df3[df3['Department'] == 'Electronics and Electrical Engineering']
cs_df= df3[df3['Department'] == 'Computer Science and Engineering']
ec_df= df3[df3['Department'] == 'Electronics and Communication Engineering']
ck_scores = [x['CodeKata Score'].sum() for x in [ee_df,ec_df,cs_df]]
```

```python
plt.pie(ck_scores,labels=departments)
plt.axis('equal')
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
topper_list = [x['CodeKata Score'].max() for x in [cs_df,ec_df,ee_df]]
plt.barh(departments, topper_list,align = 'center',height = 0.3, color = 'red')
plt.xlabel("Score")
plt.ylabel("Department")
plt.show()
```
